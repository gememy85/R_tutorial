igraph 패키지

```r
library(igraph)
```

```r
g <- graph(c(1,2, 1,3, 2,3, 3,5), n=5)
```

```r
plot(g)
```

```r
g <- graph.tree(40, 4)
```

```r
plot(g)
# 이렇게 네트워크를 만들 수 있다.
```

```r
plot(g, layout=layout.fruchterman.reingold)
```

```r
tkplot(g, layout=layout.kamada.kawai)
# 새로운 창을 뜨게 해서 그래프 모델을 manipulate하게 해준다. 
```

```r
l <- layout.kamada.kawai(g)
```

```r
rglplot(g, layout=l)
# rglplot을 이용해서도 그래프를 3d로 볼 수 있게끔해준다. 구조를 훨씬더 잘 이해할 수 있다.
```

```r
plot(g, layout=l, vertex.color="cyan")
```

```r
jets <- read.table("http://www.jaredlander.com/data/routes.csv", sep=",", header=TRUE, stringsAsFactors = FALSE)
```

```r
jets %>% head
```

```r
flights <- graph.data.frame(jets, directed=TRUE)
```

```r
print.igraph(flights, full=TRUE)
```

```r
plot(flights)
```

```r
V(flights)
```

```r
plot(flights, layout=layout.fruchterman.reingold, edge.width=E(flights)$Time/100)
```

```r
plot(flights, layout=layout.kamada.kawai)
```

```r
average.path.length(flights)
diameter(flights)
farthest.nodes(flights)
```

```r
V(flights)[farthest.nodes(flights)]
```

```r
largest.cliques(flights)
```
