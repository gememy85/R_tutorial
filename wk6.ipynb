tbl을 사용해보자
tbl은 저장하거나 보관하는데 편리하다.

```r
data(diamonds, package="ggplot2")
```

```r
class(diamonds)
```

```r
library(dplyr)
```

```r
dia <- as.tbl(diamonds)
# tbl은 데이터 프레임이지만 좀더 편리한 데이터 프레임이다.
```

```r
dia[, c('carat', 'price')]
```

```r
dia %>% select(carat, price)
```

```r
dia %>% select_("carat", "price") %>% head
dia %>% select(1, 7)%>% head # 1, 7번 컬럼
dia %>% select(-c(1,7)) %>% head # 1, 7컬럼 제외
dia %>% select(-carat, -price) %>% head
```

```r
# filter를 이용해보자. 
dia[dia$cut == "Ideal",] %>% head
```

```r
dia %>% filter(cut == "Ideal") %>% head # 위와 동일한 결과
```

```r
dia %>% filter(cut == "Ideal", color == "E") %>% head
```

```r
dia %>% filter(cut == "Ideal", color == "E") %>% head
dia %>% filter(cut == "Ideal"& color == "E") %>% head # 결과가 동일하다. 
```

```r
dia %>% filter(cut == "Ideal" | cut == "Premium") %>% head
dia %>% filter(cut %in% c("Ideal", "Premium")) %>% head
```

```r
dia %>% filter_("cut=='Ideal'") %>% head 
# 이렇게 하는 것을 우리가 이미 기존에 선언해놓은 변수가 있을 수 있기 때문이다.
dia %>% filter_(~cut=="Ideal") %>% head # 이렇게 할 수도 있겠다. 
```
